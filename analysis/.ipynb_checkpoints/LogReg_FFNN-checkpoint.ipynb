# Rice Image Classification

## Imports

```python
import numpy as np 
import pandas as pd 
import tensorflow as tf 
import keras_tuner as kt
import sklearn 
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

## Working with Predetermined Feature Set

### Data Ingestion

```python
df = pd.read_excel('./Rice_MSC_Dataset/Rice_MSC_Dataset.xlsx')
df.head(5)
```

```python
# basic facts about the data 

print(f'Number of Columns: {df.shape[1]}')
print(f'Number of Rows: {df.shape[0]}')
```

### Feature Engineering

```python
# shuffling 

np.random.seed(0) # putting seed here so in same cell as the random function 

indices = list(df.index)
shuffled_indices = np.random.permutation(indices)

df = df.reindex(shuffled_indices)
df.head(5)
```

```python
# train-test split 60-20-20 

X = df.drop(columns = ['CLASS'], axis = 1)
Y = df[['CLASS']]
Y['CLASS'] = pd.factorize(Y['CLASS'])[0] # assign the class names numeric values

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 1234) # want 80% train for later 60-20 train-val split
X_train, X_val, Y_train, Y_val = train_test_split(X_train, Y_train, test_size = 0.25, random_state = 1234) # 0.25 * .80 = 0.2
```

```python
# remove missing values since number of missing values is so small 

print(f' Number of missing feature values in training data: {sum(X_train.isna().sum())}')
print(f' Number of missing outcome values in training data: {Y_train.isna().sum()[0]}')
print(f' Number of missing feature values in test data: {sum(X_test.isna().sum())}')
print(f' Number of missing outcome values in test data: {Y_test.isna().sum()[0]}')
print(f' Number of missing feature values in validation data: {sum(X_val.isna().sum())}')
print(f' Number of missing outcome values in validation data: {Y_val.isna().sum()[0]}')

X_train.dropna()
X_val.dropna()

Y_train = Y_train[Y_train.index.isin(X_train.index)]
Y_val = Y_val[Y_val.index.isin(X_val.index)]
```

```python
# feature selection - getting rid of redundant features

# want to get rid of high multicollinearity (r > 0.8) 

X_train_corr = X_train.corr()
multicoll_mat = X_train_corr[(X_train_corr.abs() >= 0.8) & (X_train_corr.abs() < 1)] # every column has r = 1 w/ itself 

multicoll_cols = set()

for col in multicoll_mat.columns:
    correlated_columns = multicoll_mat[col].dropna().index
    multicoll_cols.update(correlated_columns)   

X_train = X_train.drop(columns = list(multicoll_cols), axis = 1)
X_test = X_test.drop(columns = list(multicoll_cols), axis = 1)
X_val = X_val.drop(columns = list(multicoll_cols), axis = 1)

print(multicoll_cols)
print(f'Number of Remaining Columns: {len(df.columns-1) - len(multicoll_cols)}')
```

```python
# do our features have predictive power? 

X_train_0 = X_train[Y_train.values == 0]
X_train_1 = X_train[Y_train.values == 1]
X_train_2 = X_train[Y_train.values == 2]
X_train_3 = X_train[Y_train.values == 3]
X_train_4 = X_train[Y_train.values == 4] # splitting X by class 

# creating histograms to see if features show class separation 

fig, ax = plt.subplots((int(len(X_train.columns)) + 1)//3, 3, figsize = (50, 20))
row_counter = 0 # for tracking axes for the plots
col_counter = 0 
for column in X_train.columns:
    ax[row_counter][col_counter].hist(X_train_0[column], label = 'Class 0', color = 'blue', alpha = 0.5)
    ax[row_counter][col_counter].hist(X_train_1[column], label = 'Class 1', color = 'purple', alpha = 0.5)
    ax[row_counter][col_counter].hist(X_train_2[column], label = 'Class 2', color = 'red', alpha = 0.5)
    ax[row_counter][col_counter].hist(X_train_2[column], label = 'Class 3', color = 'orange', alpha = 0.5)
    ax[row_counter][col_counter].hist(X_train_2[column], label = 'Class 4', color = 'green', alpha = 0.5)
    ax[row_counter][col_counter].set_title(f'{column} Frequency')
    ax[row_counter][col_counter].set_xlabel(f'{column}')
    ax[row_counter][col_counter].set_ylabel('Frequency')
    ax[row_counter][col_counter].legend()
    col_counter += 1
    if col_counter == 3:
        col_counter = 0
        row_counter += 1
plt.tight_layout() # ensures no overlap of titles 
plt.show();
```

```python
# normalizing features for easier scaling - no other transformation needed since all variables are numeric 

X_train_norm = (X_train - X_train.min()) / (X_train.max() - X_train.min())
X_test_norm = (X_test - X_test.min()) / (X_test.max() - X_test.min())
X_val_norm = (X_val - X_val.min()) / (X_val.max() - X_val.min())
```

### Baseline Model

```python
# using majority class model 

majority_class = Y_train.value_counts().idxmax()[0] # finds the majority class 
majority_class_pred = np.full(len(Y_train), majority_class) # np.full fills an array of values (majority_class) at a given length 
majority_class_accuracy = np.mean(majority_class_pred == Y_train.values) # flatten values to get same dimensions
print(f'Majority Class: {majority_class}')
print(f'Majority Class Baseline Accuracy: {majority_class_accuracy}')
```

```python
# why low accuracy? We have good class balance 

Y_train.value_counts()
```

```python
# include CE loss for the baseline model for completeness

def multiclass_cross_entropy(y, y_pred):
    """
    Computes categorical cross-entropy loss.

    Args:
    y: NumPy array of true labels (one-hot encoded).
    y_pred: NumPy array of predicted probabilities.

    Returns:
    Multi-class cross-entropy loss.
    """
    
    y = np.array(y)
    y_pred = np.array(y_pred)
    y_pred = np.clip(y_pred, 1e-15, 1 - 1e-15) # ensures no log(0) values 
    loss = -np.sum(y * np.log(y_pred))
    return loss

baseline_loss = multiclass_cross_entropy(Y_train, majority_class_pred)
print(f'Baseline Model Loss: {baseline_loss}')
```

### Hyperparameter-Tuned Multiclass Predictor 

```python
def build_logreg_model(learning_rate, reg_strength):
    """Return a simple logistic regression model using the Keras Sequential API.

        Args:
            learning_rate: The desired learning rate for SGD.
            reg_strength: strength of regularization factor 

        Returns:
            model: A tf.keras model (graph).
    """
  # Clear session and set a random seed for consistent behavior.
    tf.keras.backend.clear_session()
    tf.random.set_seed(1234)
    

  # Use Keras Sequential API to build a logistic regression
    model = tf.keras.Sequential()
    
    # create input layer
    model.add(tf.keras.Input(shape=(X_train.shape[1],), name='Input'))

    # create output layer
    model.add(tf.keras.layers.Dense(
          units=Y_train['CLASS'].nunique(),
          use_bias=True,
          activation='softmax',
          kernel_initializer=tf.keras.initializers.RandomNormal(stddev=0.01),
          bias_initializer=tf.keras.initializers.RandomNormal(stddev=0.01),
          kernel_regularizer = tf.keras.regularizers.l1_l2(l1 = reg_strength/2, l2 = reg_strength/2), # use elastic net regularization
          name='Output'
    ))
  
  # Use Categorical Crossentropy as our loss and the Adam optimizer.
    model.compile(
        loss=tf.keras.losses.SparseCategoricalCrossentropy(),
        optimizer=tf.keras.optimizers.Adam(learning_rate=learning_rate, clipnorm = 1.0), # Add gradient clipping to prevent exploding gradients
        metrics=['accuracy']
    )
    return model
```

```python
def build_logreg_model_tuner(hp):
    """Return a simple logistic regression model using the Keras Sequential API and hyperparameter tuning.

        Args:
            hp: A HyperparamModel object that allows us to perform hyperparameter tuning
            

        Returns:
            model: A tf.keras model (graph).
    """
  # Clear session and set a random seed for consistent behavior.
    tf.keras.backend.clear_session()
    tf.random.set_seed(1234)

    # define batch size for search 
    hp.Int('batch_size', min_value=16, max_value=64, step=16)
    

    # Define learning rate with more conservative options
    hp_learning_rate = hp.Float("learning_rate", min_value=1e-10, max_value=1e-5, sampling="log")

    # set parameters for regularization 
    reg_strength = hp.Float('regularization_strength', 
                               min_value=1e-5, 
                               max_value=1e-2, 
                               sampling='log')

    model = build_logreg_model(learning_rate=hp_learning_rate, reg_strength=reg_strength)

    return model 
```

```python
def hyperparam_tune(model_tuner, batch_sizes, fit_epochs):
    ''' Performs hyperpameter search for optimal parameters 
    Args: 
        model_tuner: hp model that is used to perform the search 
        batch_sizes: A list of batch sizes to hyperparameter tune for 
        fit_epochs: Number of epochs to fit model on for accuracy evaluation 

    Returns: 
        best_hps: optimal set of hyperparameters
        best_lr: optimal learning rate
        best_epoch: optimal number of epochs 
        best_batch_size: optimal number of batches 

    '''

    tf.random.set_seed(0)

    best_val_loss = float('inf')
    best_batch_size = None
    best_lr = None
    best_model = None
    
    # For each batch size, run a separate hyperparameter search
    for batch_size in batch_sizes:
        print(f"Tuning with batch_size = {batch_size}")
        
     
        # Create a tuner instance for this batch size
        tuner = kt.Hyperband(model_tuner,
            objective='val_loss',
            max_epochs=30, # add more epochs for tuning 
            factor=3,
            max_consecutive_failed_trials = 5, # allow for more failed trials 
            directory=f'tuner_dir_{batch_size}',  # Separate directory for each run
            project_name='learning_rate_tuning'
        )
        
        # Run the search with this fixed batch size
        tuner.search(
            x=X_train_norm,
            y=Y_train,
            epochs=100,
            validation_data=(X_val_norm, Y_val),
            batch_size=batch_size,  # Fixed batch size for this run
            callbacks=[
            tf.keras.callbacks.EarlyStopping( # stops the training if loss doesn't improve
                monitor='val_loss',  # Monitor loss instead of accuracy
                patience=5, # want decently high patience for 100 epochs 
                restore_best_weights=True,
                mode='min'
            ),
            tf.keras.callbacks.ReduceLROnPlateau( # drops learning rate if loss doesn't improve
                monitor='val_loss',
                factor=0.5,
                patience=5, 
                min_lr=1e-10
            )
            ],
            verbose=1
        )
        
        # Get best hyperparameters for this batch size
        best_hps = tuner.get_best_hyperparameters(num_trials=1)[0]
        current_lr = best_hps.get('learning_rate')
        
        # Evaluate the best model for this batch size
        model = tuner.hypermodel.build(best_hps)
        history = model.fit(
            x=X_train_norm,
            y=Y_train,
            epochs=100,  # Short evaluation run
            validation_data=(X_val_norm, Y_val),
            batch_size=batch_size,
            verbose=0
        )
        
        # Get the best validation loss
        current_val_loss = min(history.history['val_loss'])
        
        # If this batch size gives better results, save it
        if current_val_loss < best_val_loss:
            best_val_loss = current_val_loss
            best_batch_size = batch_size
            best_lr = current_lr
            best_model = model
    
    
    history = best_model.fit(x = X_train_norm, 
                        y = Y_train, 
                        epochs = fit_epochs,
                        validation_data = (X_val_norm, Y_val), 
                        batch_size = best_batch_size,
                        callbacks=[
                        tf.keras.callbacks.EarlyStopping( # stops the training if loss doesn't improve
                            monitor='val_loss',  # Monitor loss instead of accuracy
                            patience=10, # want relatively high patience
                            restore_best_weights=True,
                            mode='min'
                        ),
                        tf.keras.callbacks.ReduceLROnPlateau( # drops learning rate if loss doesn't improve
                        monitor='val_loss',
                        factor=0.5,
                        patience=10,
                        min_lr=1e-10
                        )
                        ],
                        verbose = 0) # avoiding printing all epochs b/c will take up too much space 
    
    val_loss_per_epoch = history.history['val_loss']
    best_epoch = val_loss_per_epoch.index(min(val_loss_per_epoch)) + 1 # find epoch that minimizes val loss

    return best_hps, best_lr, best_epoch, best_batch_size
```

```python
# perform tuning 

best_hps, best_lr, best_epoch, best_batch_size = hyperparam_tune(build_logreg_model_tuner, [32, 64], 200)
best_reg_strength = best_hps.get('regularization_strength')

# Build and compile model

model_hp = build_logreg_model(
    learning_rate=best_lr,
    reg_strength=best_reg_strength) # training on the optimal lr, regularization parameter 


# Fit the model
    
baseline_fit_hp = model_hp.fit(
    x = X_train_norm,
    y = Y_train,
    validation_data = (X_val_norm, Y_val),
    epochs = best_epoch,
    batch_size = best_batch_size,
    verbose = 0) 

print(f"Optimal Learning Rate: {best_lr}")
print(f'Optimal Number of Epochs: {best_epoch}')
print(f'Optimal Batch Size: {best_batch_size}')
print(f'Optimal Regularization Parameter: {best_reg_strength}')
print(f"Validation Accuracy: {baseline_fit_hp.history['val_accuracy'][best_epoch-1]}")
print(f"Training Accuracy: {baseline_fit_hp.history['accuracy'][best_epoch-1]}")
```

```python
# plot the loss

plt.figure(figsize = (50, 15))
plt.plot(baseline_fit_hp.history['loss'], label = 'Training Loss') # have to use .history to access information
plt.plot(baseline_fit_hp.history['val_loss'], label = 'Validation Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.title('Training and Validation Loss of Multiclass Logistic Regression Model')
plt.legend()
plt.xticks(np.arange(200) + 1)
plt.show();
```

```python
# plot the accuracy

plt.figure(figsize = (40, 15))
plt.plot(baseline_fit_hp.history['accuracy'], label = 'Training Accuracy') # have to use .history to access information
plt.plot(baseline_fit_hp.history['val_accuracy'], label = 'Validation Accuracy')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.title('Training and Validation Accuracy of Multiclass Logistic Regression Model')
plt.legend()
plt.xticks(np.arange(200) + 1)
plt.show();
```

### Hyperparameter-Tuned Feed Forward Neural Network 

```python
def build_ffnn_model(
                hidden_layer_sizes,
                dropout_rates,
                learning_rate,
                reg_strength):
    """Build a multi-class FFNN model using Keras.

    Args:
    hidden_layer_sizes: Number of neurons in each hidden layer 
    dropout_rates: percentage of neurons to be turned off in each layer
    learning_rate: The desired learning rate for the optimizer.
    reg_strength: strength of regularization factor

    Returns:
    model: A tf.keras model (graph).
    """
    tf.keras.backend.clear_session()
    np.random.seed(0)
    tf.random.set_seed(0)

    # Use Keras Sequential API to build a logistic regression
    model = tf.keras.Sequential()
    
    # create input layer
    model.add(tf.keras.Input(shape=(X_train.shape[1],), name='Input')) # both fashion and digits training have same number of obs

    # create hidden layers
    layer_number = 1 # created to help ensure names are unique
    for i in range(len(hidden_layer_sizes)):
        model.add(tf.keras.layers.Dense(
              units=hidden_layer_sizes[i],
              activation='relu',
              name = f'Hidden_{layer_number}'
        ))
        model.add(tf.keras.layers.Dropout(
            rate=dropout_rates[i],
            name = f'Dropout{layer_number}'
        )) 
        model.add(tf.keras.layers.GaussianNoise(0.1)) # adds in noise to further prevent overfitting
        layer_number += 1
         

    # create output layer
    model.add(tf.keras.layers.Dense(
          units=Y_train['CLASS'].nunique(),
          use_bias=True,
          activation='softmax',
          kernel_initializer=tf.keras.initializers.RandomNormal(stddev=0.01),
          bias_initializer=tf.keras.initializers.RandomNormal(stddev=0.01),
          kernel_regularizer = tf.keras.regularizers.l1_l2(l1 = reg_strength/2, l2 = reg_strength/2),
          name='Output'
    ))
  
  # Use Categorical Crossentropy as our loss and the SGD optimizer.
    model.compile(
        loss=tf.keras.losses.SparseCategoricalCrossentropy(),
        optimizer=tf.keras.optimizers.Adam(learning_rate=learning_rate, clipnorm = 1.0),
        metrics=['accuracy']
    )
    return model
```

```python
def build_ffnn_model_tuner(hp):
    """Build a multi-class FFNN model using Keras and Hyperparameter tuning.

    Args:
        

    Returns:
        model: A tf.keras model (graph).
        
    """
    tf.keras.backend.clear_session()
    tf.random.set_seed(0)

    # Tune number of hidden and dropout layers (1-3)
    num_hidden_layers = hp.Int('num_hidden_layers', min_value=1, max_value=3)

    hidden_layer_sizes = []
    dropout_rates = []
    
    for i in range(num_hidden_layers):
        # Tune units in each layer, will have one dropout for every hidden layer 
        units = hp.Int(f'units_layer_{i}', min_value=32, max_value=256, step=32)
        rate = hp.Float(f'rate_{i}', min_value=0.0, max_value=0.5, step=0.1)
        
        hidden_layer_sizes.append(units)
        dropout_rates.append(rate)

    # set parameters for regularization 
    reg_strength = hp.Float('regularization_strength', 
                               min_value=1e-5, 
                               max_value=1e-1, 
                               sampling='log')
    
    
    # Define learning rate with more conservative options
    hp_learning_rate = hp.Float("learning_rate", min_value=1e-10, max_value=1e-5, sampling="log")

    model = build_ffnn_model(hidden_layer_sizes, dropout_rates, hp_learning_rate, reg_strength)
  
    return model
```

```python
# perform tuning 

best_hps, best_lr, best_epoch, best_batch_size = hyperparam_tune(build_ffnn_model_tuner, [32, 64], 400)
best_reg_strength = best_hps.get('regularization_strength')

# Build and compile model

# Extract number of hidden layers
num_hidden_layers = best_hps.get('num_hidden_layers')

# Get the sizes of each hidden layer, dropout rates of each dropout layer
best_hidden_layers = []
best_dropout_rates = []
for i in range(num_hidden_layers):
    layer_size = best_hps.get(f'units_layer_{i}')
    rate = best_hps.get(f'rate_{i}')
    
    best_hidden_layers.append(layer_size)
    best_dropout_rates.append(rate)

model_hp = build_ffnn_model( 
                            hidden_layer_sizes = best_hidden_layers,
                            dropout_rates = best_dropout_rates,
                            learning_rate=best_lr,
                            reg_strength=best_reg_strength) # training on the optimal params 


# Fit the model
    
baseline_fit_hp = model_hp.fit(
    x = X_train_norm,
    y = Y_train,
    validation_data = (X_val_norm, Y_val),
    epochs = best_epoch,
    batch_size = best_batch_size,
    verbose = 0) 

print(f"Optimal Learning Rate: {best_lr}")
print(f'Optimal Number of Epochs: {best_epoch}')
print(f'Optimal Batch Size: {best_batch_size}')
print(f'Optimal Regularization Strength: {best_reg_strength}')
print(f'Optimal Hidden Layer Sizes: {best_hidden_layers}')
print(f'Optimal Dropout Rates: {best_dropout_rates}')
print(f"Validation Accuracy: {baseline_fit_hp.history['val_accuracy'][best_epoch-1]}")
print(f"Training Accuracy: {baseline_fit_hp.history['accuracy'][best_epoch-1]}")
```

```python
# plot the loss

plt.figure(figsize = (50, 15))
plt.plot(baseline_fit_hp.history['loss'], label = 'Training Loss') # have to use .history to access information
plt.plot(baseline_fit_hp.history['val_loss'], label = 'Validation Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.title('Training and Validation Loss of Multiclass Logistic Regression Model')
plt.legend()
plt.xticks(np.arange(400) + 1)
plt.show();
```

```python
# plot the accuracy

plt.figure(figsize = (30, 10))
plt.plot(baseline_fit_hp.history['accuracy'], label = 'Training Accuracy') # have to use .history to access information
plt.plot(baseline_fit_hp.history['val_accuracy'], label = 'Validation Accuracy')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.title('Training and Validation Accuracy of Multiclass Logistic Regression Model')
plt.legend()
plt.xticks(np.arange(400) + 1)
plt.show();
```
